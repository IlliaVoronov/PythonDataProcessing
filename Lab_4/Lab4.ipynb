<h3>Task 1</h3>

```python
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt

# 1. Load the image
image_path = '12.jpg'
img = Image.open(image_path)

# 2. Convert the image to grayscale
gray_img = img.convert('L')

# 3. Apply edge detection
edges = gray_img.filter(ImageFilter.FIND_EDGES)

# 4. Invert the colors
inverted_edges = edges.point(lambda x: 255 - x)

gray_img.save('/Users/illiavoronov/Documents/UNI/Python Data Processing/Lab_4/result_images/gray_img.jpg')
inverted_edges.save('/Users/illiavoronov/Documents/UNI/Python Data Processing/Lab_4/result_images/inverted_edges_image.jpg')

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(gray_img, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')


plt.subplot(1, 2, 2)
plt.imshow(inverted_edges, cmap='gray')
plt.title('Inverted Edge Detection')
plt.axis('off')
```

<h3>Task 2</h3>

```python
import random
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt

# 1. Load and resize the image
image_path = 'Harry.png'
img = Image.open(image_path)
img = img.resize((500, 500))

# 2. shuffle and create Warhol-style images
def create_warhol_poster(img, poster_name):
    channels = img.split()

    poster = Image.new('RGB', (1500, 1500))

    for i in range(3):
        for j in range(3):
            random_channels = random.sample(channels, 2)  # Pick two channels randomly
            new_image = Image.merge("RGB", (random_channels[0], random_channels[1], random_channels[0]))
            poster.paste(new_image, (i * 500, j * 500))  # Place it in the grid

            # 3. Add captions
            draw = ImageDraw.Draw(poster)
            caption = f"Channels: {random_channels[0].mode}, {random_channels[1].mode}"
            draw.text((i * 500 + 20, j * 500 + 450), caption, fill="white")

    poster.save(f'result_images/{poster_name}.jpg')

    # Display the poster
    plt.figure(figsize=(10, 10))
    plt.imshow(poster)
    plt.axis('off')
    plt.title(poster_name)
    plt.show()


create_warhol_poster(img, 'poster1')
create_warhol_poster(img, 'poster2')
create_warhol_poster(img, 'poster3')
```
