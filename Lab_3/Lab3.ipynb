```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset df - data frame)
df = pd.read_csv('NationalNames.csv')
```


<h4>6. Calculate the number of unique female and male names in whole dataset</h4>

```python

unique_female_names = df[df['Gender'] == 'F']['Name'].nunique()
unique_male_names = df[df['Gender'] == 'M']['Name'].nunique()

print(f"Number of unique female names: {unique_female_names}")
print(f"Number of unique male names: {unique_male_names}")
```

<h4>7. Find 5 the most popular male names in 2010</h4>


```python
male_2010 = df[(df['Year'] == 2010) & (df['Gender'] == 'M')]

top_5_male_names_2010 = male_2010.groupby('Name')['Count'].sum().sort_values(ascending=False).head(5)

print(top_5_male_names_2010)
```

<h4>8. Find the most popular name based on the results of one year (the name for which
Count is maximum)</h4>

```python
most_popular_name = df.loc[df['Count'].idxmax()]

print(most_popular_name)
```

<h4>13. Find the year when the name “Jacob” was the most popular as a female name</h4>

```python
jacob_female = df[(df['Name'] == 'Jacob') & (df['Gender'] == 'F')]

if not jacob_female.empty:
    jacob_female_max_year = jacob_female.loc[jacob_female['Count'].idxmax()]
    print(jacob_female_max_year)
else:
    print("No records of 'Jacob' as a female name.")
```

<h4>14. Find year, with the most number of gender neutral names (the same male and
female names)</h4>

```python
gender_neutral_names = df.groupby(['Year', 'Name', 'Gender']).size().unstack().dropna()
neutral_names_count = gender_neutral_names.groupby('Year').size()
year_max_neutral = neutral_names_count.idxmax()

print(f"Year with most gender-neutral names: {year_max_neutral}")
```

<h4>15. Find total births per year</h4>

```python
total_births_per_year = df.groupby('Year')['Count'].sum()

print(total_births_per_year.head())
```

<h4>16. Find the year when the greatest number of children was born</h4>

```python
year_max_births = total_births_per_year.idxmax()

print(f"Year with the greatest number of children born: {year_max_births}")
```

<h4>17. Find the number of girls and boys that were born in each year</h4>

```python
births_by_gender_per_year = df.groupby(['Year', 'Gender'])['Count'].sum().unstack()

print(births_by_gender_per_year.head())
```

<h4>19. Draw the plot of total births per year of boys and girls</h4>

```python
births_by_gender_per_year.plot(kind='line', title='Total births per year (boys and girls)')
plt.xlabel('Year')
plt.ylabel('Number of births')
plt.show()
```

<h4>20. Count number of gender neutral names (same for girls and boys)</h4>

```python
gender_neutral_names = df.groupby(['Name', 'Gender']).size().unstack().dropna()
total_gender_neutral_names = gender_neutral_names.shape[0]

print(f"Total number of gender-neutral names: {total_gender_neutral_names}")
```

<h4>22. Calculate how many years the observation was carried out</h4>

```python
unique_years = df['Year'].nunique()

print(f"Number of years of observation: {unique_years}")
```

<h4>24. Find the most popular unpopular names. unpopular name that babies have been
called the most times</h4>

```python
unpopular_names = df.groupby('Name')['Count'].sum().sort_values()
most_popular_unpopular = unpopular_names.idxmax()

print(f"Most popular unpopular name: {most_popular_unpopular}")
```

<h4>25. Plot graphs of the distribution of the number of names “John” and “Mary” by years, regardless of gender</h4>

```python
john_mary = df[df['Name'].isin(['John', 'Mary'])].groupby(['Year', 'Name'])['Count'].sum().unstack()

john_mary.plot(kind='line', title='Distribution of John and Mary names by year')
plt.xlabel('Year')
plt.ylabel('Number of births')
plt.show()
```

<h4>26. Plot graphs of the distribution of female names “John” and male names “Mary” by years</h4>

```python
john_female_mary_male = df[((df['Name'] == 'John') & (df['Gender'] == 'F')) | ((df['Name'] == 'Mary') & (df['Gender'] == 'M'))]

john_female_mary_male.groupby(['Year', 'Name'])['Count'].sum().unstack().plot(kind='line', title='Distribution of female John and male Mary names by year')
plt.xlabel('Year')
plt.ylabel('Number of births')
plt.show()
```
